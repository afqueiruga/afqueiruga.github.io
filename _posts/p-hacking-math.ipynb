```python
import matplotlib.pyplot as plt
```

```python
ppass = 0.7**100
p_lucky = 0.0
y_hack = []
x_hack = []
for i in range(20):
    p_lucky = p_lucky + (1.0-p_lucky)*ppass
    x_hack.append(i+1)
    y_hack.append(p_lucky)
    # if i%1_000_000==0: print(i+2, p_lucky)
plt.plot(x_hack, y_hack)
```

```python
ppasses = [0.7**5, 0.9**5, 0.95**5, 0.99**5]
for ppass in ppasses:
    p_lucky = ppass
    y = [p_lucky]
    x = [1]
    for i in range(20):
        p_lucky = p_lucky + (1.0-p_lucky)*ppass
        x.append(i+2)
        y.append(p_lucky)
    plt.plot(x, y, label=f'{ppass:.2f}')
```

```python
n_test = 100
ppasses = [0.7**n_test, 0.8**n_test, 0.9**n_test, 0.95**n_test, 0.99**n_test]
p_lucky = 0.0
x_working = []
y_working = []
for i in range(20):
    try:
        ppass = ppasses.pop(0)
    except IndexError:
        pass
    p_lucky = p_lucky + (1.0-p_lucky)*ppass
    x_working.append(i+1)
    y_working.append(p_lucky)
plt.plot(x_working, y_working, 'o-', label=f'progress')
plt.plot(x_hack, y_hack, 'o-', label='no progress')
plt.xlabel('Trials')
plt.ylabel('Probability of a perfect score.')
plt.legend()
```
